```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('score.csv')
```

```python
# Made data and split it 
from sklearn.model_selection import train_test_split

X=df[['Hours']]
y=df['Scores']

x_train,x_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=42)

```

```python
#fitting data into linear regression
from sklearn.linear_model import LinearRegression
slr= LinearRegression()
slr.fit(x_train,y_train)
```

```python
#intercept and cofficient 
print('intercept of line is',slr.intercept_)
print('coffiecient of line is',slr.coef_)
y_pred=slr.predict(x_test)
# print(y_test)
# print(y_pred)
```

```python
#make a scattre plot and best fit line
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,'red')

# plt.scatter(X,y)
# plt.xlabel('Hours')
# plt.ylabel('score')
```

```python
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred)
meanSqErr = metrics.mean_squared_error(y_test, y_pred)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('R squared: {:.2f}'.format(slr.score(X,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```
